```python
# droping rows with missing values
df.dropna(axis=0, how='any', inplace= True)
```

```python
# Number of Observation on each day
a=df['ObservationDate'].value_counts()
frame={'Date': a.index, 'Sum':a}
df1=pd.DataFrame(frame)
df1.set_index('Date', inplace = True)
df1
```

```python
#Trend of Observations
observations = df.groupby('ObservationDate')['ObservationDate'].count()
plt.plot(observations.index, observations, color = 'b', label = 'Date Vs Observations')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Observations')
plt.xticks(a.index, " ")[1]
```

```python
#Trend of Confirmed Cases
confirmed = df.groupby('ObservationDate')['Confirmed'].sum()
plt.plot(confirmed.index, confirmed, color = 'orange', label = 'Date Vs Cases')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Cases per Day')
plt.xticks(confirmed.index, " ")[1]
```

```python
#Trend of Deaths
deaths = df.groupby('ObservationDate')['Deaths'].sum()
plt.plot(deaths.index, deaths, color = 'r', label = 'Date Vs Deaths')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Deaths per Day')
plt.xticks(deaths.index, " ")[1]
```

```python
#Trend of Recovered Patients
recovered = df.groupby('ObservationDate')['Recovered'].sum()
plt.plot(recovered.index, recovered, color = 'g', label = 'Date Vs Recovered Patients')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Recovered Patients per Day')
plt.xticks(recovered.index, " ")[1]
```

```python
# All Trends in one Graph
plt.figure(figsize=(10,10))

plt.plot(observations.index, observations, color = 'b', label = 'Date Vs Observations' , lw=3)
plt.plot(confirmed.index, confirmed, color = 'orange', label = 'Date Vs Cases' , lw=3)
plt.plot(deaths.index, deaths, color = 'r', label = 'Date Vs Deaths' , lw=3)
plt.plot(recovered.index, recovered, color = 'g', label = 'Date Vs Recovered Patients' , lw=3)
plt.legend()
plt.xlabel('Date')
plt.xticks(recovered.index, '')[1]
```

# Death and Recovery Percentage

```python
death_sum = df.groupby('Country/Region')['Deaths'].sum()
confirmed_sum = df.groupby('Country/Region')['Confirmed'].sum()
recovered_sum = df.groupby('Country/Region')['Recovered'].sum()
 
d_percentages = {'Confirmed Sum' : confirmed_sum, 
                'Death Sum' : death_sum, 
                'Recovered Sum' : recovered_sum}
 
df_percentages = pd.DataFrame(d_percentages)
df_percentages['Death %'] = (df_percentages['Death Sum'] / df_percentages['Confirmed Sum']) * 100
df_percentages['Recover %'] = (df_percentages['Recovered Sum'] / df_percentages['Confirmed Sum']) * 100
df_percentages.sort_values('Recover %', ascending = False)
```
